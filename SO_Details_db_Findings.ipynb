```python
#converting series to dataframe and resetting index, and changing the column name to 'Orders'
df_orders_group=pd.DataFrame(df_orders_group)
df_orders_group=df_orders_group.reset_index()
df_orders_group=df_orders_group.rename(columns={"CURQTY_28":"Orders"})
```

```python
df_orders_group.head()
```

```python
plot_df(df, x=df_orders_group.Date, y=df_orders_group.Orders,title='Orders from 2014 to 2021.')
```

```python
df_orders_group['Year'] = [d.year for d in df_orders_group.Date]
df_orders_group['Month'] = [d.strftime('%b') for d in df_orders_group.Date]
years = df_orders_group['Year'].unique()
```

```python
df_orders_group.head()
```

```python
#Colors
np.random.seed(100)
mycolors = np.random.choice(list(mpl.colors.XKCD_COLORS.keys()), len(years), replace=False)
```

```python
plt.figure(figsize=(16,8), dpi= 100)
for i, y in enumerate(years):
    if i >= 0:        
        plt.plot('Month', 'Orders', data=df_orders_group.loc[df_orders_group.Year==y, :], color=mycolors[i], label=y)
        plt.text(df_orders_group.loc[df_orders_group.Year==y, :].shape[0]-.9, df_orders_group.loc[df_orders_group.Year==y, 'Orders'][-1:].values[0], y, fontsize=12, color=mycolors[i])


# Decoration
plt.gca().set(ylabel='$Number-of-Orders$', xlabel='$Month$')
plt.yticks(fontsize=12, alpha=.7)
plt.title("Plot of Orders Time Series", fontsize=20)
plt.show()
```

```python
#Month-wise trend
sns.boxplot(x='Month', y='Orders', data=df_orders_group)
```
